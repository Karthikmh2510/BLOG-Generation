# Prompt Template Text Summarization

```python
from langchain.chains import LLMChain
from langchain import PromptTemplate
```

```python
generic_template = """
Write a summary of the following speech:
Speech : '{speech}'
Translate the precise summary to {language}
"""

prompt = PromptTemplate(
    input_variables = ['speech','language'],
    template = generic_template
)
```

```python
complete_prompt = prompt.format(speech=speech, language='Kannada')
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm.invoke(complete_prompt)
```

```python
print(llm.invoke(complete_prompt).content)
```

```python
llm_chain = LLMChain(llm=llm, prompt=prompt)
summary = llm_chain.invoke({'speech':speech, 'language':'Hindi'})
```

```python
print(summary)
```

```python
print(summary['text'])
```

# StuffDocumentChain Techniques for text Summarixation

```python
from PyPDF2 import PdfReader
```

```python
# Provide the path of the pdf
pdfreader = PdfReader('apjspeech.pdf')
pdfreader
```

```python
from typing_extensions import Concatenate

text =''
for i, page in enumerate(pdfreader.pages):
    content = page.extract_text()
    if content:
        text += content
```

```python
text
```

```python
# We need to convert all text into docs
from langchain.docstore.document import Document
docs = [Document(page_content=text)]
docs
```

```python
llm = ChatOpenAI(model='gpt-3.5-turbo', temperature=0.1)
```

```python
llm.get_num_tokens(text)
```

```python
from langchain.chains.summarize import load_summarize_chain
template = '''
Write a concise and short summary of the following speech.
SPeech : '{text}'
'''

prompt = PromptTemplate(
    input_variables=['text'],
    template=template
)
```

```python
chain = load_summarize_chain(
    llm,
    chain_type='stuff', #stuff means chain is waiting for the docs to come
                        # There after it will call llm to do task.
    prompt=prompt,
    verbose=True
)
```

```python
output_summary = chain.invoke(docs)
```

If we have docs of huge size, we have more tokens, so stuff reduce techniques won't work.

So, we use Map reduce techniques. Here we divide pdf into chunk, and provide every chunk to llm and get summary. And finally, all summaries are combined by llm and llm gives final summaries.

# Summarization Large docs using Map Reduce Techniques

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
llm.get_num_tokens(text)
```

```python
# Let's split these tokens into chunks using RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=8000,
    chunk_overlap=20)
chunks = text_splitter.create_documents([text])
```

```python
chunks
```

```python
chain_map_reduce = load_summarize_chain(
    llm,
    chain_type='map_reduce',
    verbose=True
)
sumary_map= chain_map_reduce.invoke(chunks)
```

```python
print(sumary_map)
```

# Map reduce with Custom Prompts

```python
chunk_prompt = """
Please summarize the below speech:
Speech: '{text}'
Summary:"
"""

map_prompt_template = PromptTemplate(
    input_variables=['text'],
    template=chunk_prompt
)
```

```python
final_combined_prompt = '''
Provide a final summary of the entire speech with these important points.
Add a Generic Motivational Title,
Start the precise summary with am introduction and provide the summary 
in number points for the speech.
Speech: '{text}'
'''

final_combined_prompt_template = PromptTemplate(
    input_variables=['text'],
    template=final_combined_prompt
)
```

```python
summary_map_custom_prompt = load_summarize_chain(
    llm,
    chain_type='map_reduce',
    map_prompt=map_prompt_template,
    combine_prompt=final_combined_prompt_template,
    verbose=False
)
```

```python
output_map_custome = summary_map_custom_prompt.invoke(chunks)
```

```python
print(output_map_custome)
```

# RefineChain for Summarization

Similar to map reduce. 

1st we pass chunk1 and get summary
2nd we pass chunk 1,2 and get summary,
3rd we pass chunk 1,2,3 and get summary...
and it goes on,

```python
chain_refine = load_summarize_chain(
    llm,
    chain_type='refine',
    verbose=True
)

output_refine = chain_refine.invoke(chunks)
```

# RefineChain for Summarization using custome prompt

```python
chunk_prompt_refinechain = """
Please summarize the below speech:
Speech: '{text}'
Summary:"
"""

refinechain_prompt_template = PromptTemplate(
    input_variables=['text'],
    template=chunk_prompt_refinechain
)

final_combined_refine_prompt = '''
Provide a final summary of the entire speech with these important points.
Add a Generic Motivational Title,
Start the precise summary with am introduction and provide the summary 
in number points for the speech.
Speech: '{text}'
'''

final_combined_prompt_template_refine = PromptTemplate(
    input_variables=['text'],
    template=final_combined_refine_prompt
)

summary_refinechain_custom_prompt = load_summarize_chain(
    llm,
    chain_type='map_reduce',
    map_prompt=map_prompt_template,
    combine_prompt=final_combined_prompt_template_refine,
    verbose=False
)

output_custome_refine = summary_refinechain_custom_prompt.invoke(chunks)
```

```python
print(output_custome_refine)
```
